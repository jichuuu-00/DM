- A two-dimensional labeled data structure with 2 or more columns

### 2-1) <u>C</u>RUD - Create

- c.f., dictionary 

```python
df = pd.DataFrame({
    "Part_ID": [137, 145, 150, 152],
    "Part_Name": ['Door latch', 'Side Mirror', 'Door molding', 'Compressor'],
    "Unit_Price": [22.0, 12.0, 6.0, 54.0],
    "Supplier_Num": [8259, 8261, 8263, 8444]},
)
df
```

```python
df = df.set_index('Part_ID')
df
```

### 2-2) C<u>R</u>UD - Read
- Read Index
- Read Columns
- Read Values

### 2-3) CR<u>U</u>D - Update
- Add new columns

```python
df['Supplier_State'] = ['NY', 'IN', 'NC', 'CA']
df
```

- Add new rows

```python
df.loc[160] = ['Door handle', 10.0, 8259, 'IL']
df
```

### 2-4) CRU<u>D</u> - Delete
- Delete columns

```python
del df['Supplier_State']
df
```

- delete rows

```python
df = df.drop([160])
df
```

### Copy
> A copy creates a new object which stores the reference of the original elements.

```python
df
```

```python
df_copy = df[['Part_Name', 'Unit_Price']].copy()
df_copy
```

- Changes made to the copied df do not affect the original df

```python
df_copy.columns = ['PartName', 'UnitPrice']
```
