### Exercise 2. String, If statement, Dictionary
* Create a dictionary of currency rate, and take currency (USD or CNY or JPY) from the user and convert it to KRW.

<table>
<thead>
<tr>

````python
curr_dict = {'USD':1216, 'CNY':191, 'JPY':10}
````

````python
currency = input()

if currency[4:] == 'USD':
    print(int(currency[:4])*curr_dict['USD'], 'KRW')
elif currency[4:] == 'CNY':
    print(int(currency[:4])*curr_dict['CNY'], 'KRW')
elif currency[4:] == 'JPY':
    print(int(currency[:4])*curr_dict['JPY'], 'KRW')
````

````python
emp_dict = {'Dept':['Employee','Check'], 'Marketing':['Ben','X'], 'Operation':['Sam','O'],
         'MIS':['Will','X'], 'Finance':['Sarah','X']}
````
