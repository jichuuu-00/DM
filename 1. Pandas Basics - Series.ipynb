</div>

- A one-dimentional labeled data structure

### 2-1) <u>C</u>RUD - Create

- default index starts from 0

```python
data = pd.Series([70, 60, 90])
data
```

- Index can be changed

```python
data = pd.Series([70, 60, 90], index = ['Math', 'English', 'Literature'])
data
```

### 2-2) C<u>R</u>UD - Read
- Read Index
- Read Values

```python
data.index
```

```python
data.values
```

- Values indicated by Index

```python
print(data['Math'])
print(data[0])
```

### 2-3) CR<u>U</u>D - Update

- Update Index

```python
data.index = ['DIGB101', 'DIGB102', 'DIGB405']
data
```

- Values can be updated, but not frequently used

```python
data['DIGB102'] = 65
data
```

### 2-4) CRU<u>D</u> - Delete

```python
del data['DIGB101']
data
```
