### Exercise 17. Interacting with String & If Statement
* Take social security number (SSN) as input from the user and print the place of birth.
* NOTE: the second and third of the second part of SSN indicate the place of birth. 

e.g., 950001-1081234

<table>
<thead>
<tr>
<th>Regional Code</th>
<th>Place of Birth</th>
</tr>
</thead>
<tbody>
<tr>
<td>00 ~ 08</td>
<td>Seoul</td>
</tr>
<tr>
<td>09 ~ 12</td>
<td>Pusan</td>
</tr>
</tbody>
</table>

````python
num = input('social security number')

num = num.split("-")[1]
place = int(num[1:3])
````

````python
if 0 <= place <= 8:
    print('Seoul')
elif 9 <= place <= 12:
    print('Pusan')
````
