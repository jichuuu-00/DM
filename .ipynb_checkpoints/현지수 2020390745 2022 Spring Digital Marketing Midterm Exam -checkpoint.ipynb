## Question 1. 
Print the multiplication table, but only those that are even number. (구구단을 출력하되 짝수인 경우에 한해서 출력하시오.) <br>
For example,
- 3 X 3 = 9: 9 which is an odd number should not be returned. (9는 홀수이므로 반환하지 않음.)  
- 3 X 4 = 12: 12 which is an even number should be returned. (12는 짝수이므로 반환함.) 


- Output example (출력 예)
<pre>
2 X 1 = 2
2 X 2 = 4
2 X 3 = 6
2 X 4 = 8
2 X 5 = 10
2 X 6 = 12
2 X 7 = 14
2 X 8 = 16
2 X 9 = 18
3 X 2 = 6
3 X 4 = 12
3 X 6 = 18
3 X 8 = 24
...
9 X 2 = 18
9 X 4 = 36
9 X 6 = 54
9 X 8 = 72
</pre>

````python
for i in range(2,10):
    for m in range(1,10):
        if i*m % 2 == 0:
            print(i, 'X', m, '=', i*m)
````

## Question 4. 
Define a function that takes three arguments and return the multiplication or division of first and second argument depending on the third argument. <br>
(세개의 인자를 받고, 세번째 인자에 따라 첫번째와 두번째 인자를 곱하거나 나눈 값을 반환하는 함수를 정의하시오.)

- Condition1: 
    - 1st argument: integer
    - 2nd argument: integer
    - 3rd argument: "*" or "/"


- Condition2: 
    - If the 3rd argument is "*", then return the multiplication of 1st and 2nd argument <br>
    (세번째 인자가 '*'인 경우, 첫번째 인자와 두번째 인자를 곱한 값을 반환한다.)
    - If the 3rd argument is "/", then return the division of 1st by 2nd argument <br>
    (세번째 인자가 '/'인 경우, 첫번째 인자를 두번째 인자로 나눈 값을 반환한다.)
    - NOTE: if the 2nd argument is 0, then return -1 <br>
    (단, 두번째 인자가 0인 경우에는 일괄적으로 -1을 반환한다.)


- Output example (출력 예): 
```python
example1:
data = func(2, 3, "*")
print (data)
6
```

  ```python
  example2:
  data = func(2, 0, "/")
  print (data)
  -1
  ```

````python
def func1(arg1, arg2, arg3):
    if arg3 == '*':
        return arg1 * arg2
    elif arg2 == 0:
        return -1
    elif arg3 == '/':
        return arg1 / arg2
````

````python
data = func1(2, 3, "*")
print (data)
````

````python
data = func1(2, 0, "/")
print (data)
````
