```python
def exp(arg1, arg2):
    value = arg1 ** arg2
    return value
```

```python
exp(2,4)
```

## How to use Library2

```python
from math import pow, factorial

data1 = pow(4,2)
print(data1)
data2 = factorial(4)
print(data2)
```

## How to use Library3

```python
from math import *
data1 = pow(4,2)
print(data1)
data2 = factorial(4)
print(data2)
```

## How to use Library4
