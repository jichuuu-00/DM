### Exercise 4. 
Define a func4() that takes three arguments and return the sum or multiplication of first and second argument depending on the third argument. 

```python
1st argument: integer
2nd argument: integer
3rd argument: "*" or "+"

If the 3rd argument is "*", then return the multiplication of 1st and 2nd argument 
If the 3rd argument is "+", then return the sum of 1st and 2nd argument 
If the 3rd argument is other than "*" or "+, then return 0

output example1:
data = func(2, 3, "*")
print (data)

6

output example2:
data = func(2, 3, "+")
print (data)

5

output example3:
data = func(2, 3, "-")
print (data)

0
```

````python
def func4(arg1, arg2, arg3):
    if arg3 == '*':
        return arg1 * arg2
    elif arg3 == '+':
        return arg1 + arg2
    else:
        return 0
````

````python
func4(2,3,'*')
````

### Exercise 5. 
Define a func5() that takes two arguments and return the output as below.

```python
1st argument: string
2nd argument: integer
return: a list of strings multiple times indicated by the 2nd argument
output example:
print_string("Python", 3)

["Python", "Python", "Python"]
```

````python
def func5(arg1, arg2):
    data = list()
    for index in range(arg2):
        data.append(arg1)
    return data
````
