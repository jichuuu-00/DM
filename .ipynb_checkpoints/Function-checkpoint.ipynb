## Parameter = argument = input

```python
def func3(arg):
    print('Hello!', arg)
```

```python
func3('Java')
```

```python
def func4(arg1, arg2):
    print('Hello!', arg1, '&', arg2)
```

```python
func4('Python', 'Java')
```

```python
def func5(arg1, arg2):
    return arg1+arg2
```

```python
func5(2,6)
```

```python
def func6(arg1,arg2):
    return arg1*(arg1 + arg2)
```

```python
func6(2,6)
```

## 3. Function with multiple returns

```python
def func9(arg1):
    arg2 = arg1 + 7
    return arg1, arg2
```

```python
func9(7)
```

```python
def id_check(arg):
    if arg == 'admin':
        print('invalid id: admin')
        return
    else:
        print('valid id:', arg)
```

```python
id_check('admin')
```
