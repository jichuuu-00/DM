## Loops in Loops

```python
res = requests.get('https://sy-netsci.github.io/DIGB241_CSS/')
soup=BeautifulSoup(res.content, 'html.parser')
rows = soup.select('tr')
for row in rows:
    columns = row.select('td')
    row_str=''
    for column in columns:
        row_str += '.' + column.get_text()
    print(row_str[1:])
```
