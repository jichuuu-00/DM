### 1. Read a CSV file

```python
import pandas as pd
df_confirmed = pd.read_csv('final_covid.csv')
df_confirmed.head()
```

```python
df_confirmed.shape
```

### 2. iso2 Info
>- Get ios2 info from 'UID_ISO_FIPS_LookUp_Table'
>- Pre-process 'iso2' info
    - iso2.duplicated()
    - iso2.drop_duplicates(subset='Country_Region', keep='last')
>

```python
iso2 = pd.read_csv('COVID-19-master/csse_covid_19_data/UID_ISO_FIPS_LookUp_Table.csv', encoding='utf-8-sig', keep_default_na=False, na_values='')
iso2.head()
```

```python
iso2 = iso2[['iso2', 'Country_Region']]
iso2.head()
```

```python
iso2.shape
```

```python
iso2.duplicated()
```

```python
iso2 = iso2.drop_duplicates(subset='Country_Region', keep='last')
iso2.shape
```

### 3. Merge 'df_confirmed' & 'iso2'

```python
doc_final = pd.merge(df_confirmed, iso2, how='left', on='Country_Region')
doc_final.head()
```

```python
doc_final.shape
```

>- Check NaN

```python
doc_final.isnull().sum()
```

```python
doc_final[doc_final['iso2'].isnull()]
```

>- Delete rows with NaN in 'iso2'

```python
doc_final = doc_final.dropna(subset=['iso2'])
```

```python
doc_final[doc_final['iso2'].isnull()]
```

### 4. Get Flag Image Links

>- Source: https://countryflagsapi.com/
>- e.g., KR: https://countryflagsapi.com/png/kr

```python
def create_flag_link(country_code):
    flag_link = 'https://countryflagsapi.com/png/' + country_code
    return flag_link
```

### 5. Apply 'Flag Image Links' to 'iso2' in doc_final

```python
doc_final['iso2'] = doc_final['iso2'].apply(create_flag_link)
```

```python
doc_final.head()
```

### 6. 'iso2' as a 2nd column

```python
cols = doc_final.columns.tolist()
```

>- Remove 'iso2'
>- Inser in the Index number 1

```python
cols.remove('iso2')
cols.insert(1, 'iso2') #index number 1
```

```python
doc_final = doc_final[cols]
```

```python
doc_final.head()
```

### 7. Save as a CSV file

```python
doc_final.to_csv('final_covid_flag.csv')
```
