### seaborn.heatmap parameters
Reference: https://seaborn.pydata.org/generated/seaborn.heatmap.html
- data=rectangular dataset (dataframe)
- annot: Boolean or rectangular dataset. If True, write the data value in each cell.
    - e.g., annot=True
- fmt: String formatting code to use when adding annotations
    - e.g., fmt=0.2f 
- linewidths: Width of the lines that will divide each cell.
    - e.g., linewidths=0.5
- cmap: 
    - e.g., cmap='Blues' 
    - color more options: https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html)

````python
plt.figure(figsize=(5,5)) #size of a plot
sns.heatmap(data = covid_April01.corr(), annot=True, fmt='.2f', linewidths=0.5, cmap='Blues')
````
