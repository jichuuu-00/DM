```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://news.v.daum.net/v/20220111190138086')
soup = BeautifulSoup(res.content, 'html.parser')
mydata = soup.find('h3', class_='tit_view')
mydata.get_text()
```

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://news.v.daum.net/v/20220111190138086')
soup = BeautifulSoup(res.content, 'html.parser')
mydata = soup.find('h3', class_='tit_cp')
mydata2 = soup.find('h3', attrs={'class':'tit_cp'})
print(mydata.get_text())
print(mydata2.get_text())
```

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://sy-netsci.github.io/DIGB241/')
soup = BeautifulSoup(res.content, 'html.parser')
mydata = soup.find_all('li', class_='digb')
for item in mydata:
    print(item.get_text())
```
