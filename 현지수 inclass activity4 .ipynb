```python
import requests
from bs4 import BeautifulSoup

key = '4sWBFgUY4SfGcWl7znB0%2BKxXMNz9gkzIkEbN3wd8uLe5CeJJMTgM%2BBN2lgB4MCITxG5RbAFbcRfYWXi2LKA47A%3D%3D'
params = '&airportCode=GMP&year=2016'
gov_api = 'http://openapi.airport.co.kr/service/rest/noiseMeasureService/getNoiseMeasure?serviceKey='+key+params

res = requests.get(gov_api)
soup = BeautifulSoup(res.content, 'html.parser')
mydata = soup.find_all('item')
for item in mydata:
    average = item.find('average')
    winter = item.find('month1')
    summer = item.find('month7')
    print('average:', average.get_text(), 'winter:', winter.get_text(), 'summer:', summer.get_text())
```

```python
import requests
from bs4 import BeautifulSoup
import openpyxl

key = '4sWBFgUY4SfGcWl7znB0%2BKxXMNz9gkzIkEbN3wd8uLe5CeJJMTgM%2BBN2lgB4MCITxG5RbAFbcRfYWXi2LKA47A%3D%3D'
params = '&airportCode=GMP&year=2016'
gov_api = 'http://openapi.airport.co.kr/service/rest/noiseMeasureService/getNoiseMeasure?serviceKey='+key+params

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['A'].width = 30
excel_sheet.column_dimensions['B'].width = 30
excel_sheet.column_dimensions['C'].width = 30
excel_sheet.append(['average', 'winter', 'summer'])

res = requests.get(gov_api)
soup = BeautifulSoup(res.content, 'html.parser')
mydata = soup.find_all('item')
for item in mydata:
    average = item.find('average')
    winter = item.find('month1')
    summer = item.find('month7')
    excel_sheet.append([average.get_text(), winter.get_text(), summer.get_text()])

excel_file.save('Noise.xlsx')
excel_file.close()
```
