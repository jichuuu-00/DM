```python
import requests
import openpyxl

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['B'].width = 50
excel_sheet.column_dimensions['C'].width = 80
excel_sheet.column_dimensions['D'].width = 30
excel_sheet.append(['Label', 'item Name', 'Link info', 'Mall Name'])

client_id = '0yhn9rQGesj92gcDKxHv'
client_secret = '_jGFm9sLbD'
params = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_secret}

i=0
for index in range(10):
    start_num = 1 + index*100
    naver_api = 'https://openapi.naver.com/v1/search/shop.json?query=OLEDTV&display=100&start='+str(start_num)
    res = requests.get(naver_api, headers=params)
    if res.status_code == 200:
        data = res.json()
        for i,item in enumerate(data['items']):
            i += 1
            excel_sheet.append([i,item['title'], item['link'], item['mallName']])
    else:
        print('Error Code', res.status_code)
        
excel_file.save('Naver_OLEDTV.xlsx')
excel_file.close()      
```
