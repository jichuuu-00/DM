```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
res = requests.get('https://sy-netsci.github.io/DIGB241')
soup=bs(res.content, 'html.parser')
mydata = soup.find_all('li', 'digb')
for data in mydata:
    print(data.get_text())
```

## Tags with Tags

```python
res = requests.get('https://sy-netsci.github.io/DIGB241')
soup=bs(res.content, 'html.parser')

#Scraping Academic Curriculum Section first
section = soup.find('ul', id='digb_course_list')
courses=section.find_all('li', 'digb')
for course in courses:
    print(course.get_text())
```

## Manipulating data
* strip()
* split()

```python
res = requests.get('https://sy-netsci.github.io/DIGB241')
soup=bs(res.content, 'html.parser')

#Scraping Academic Curriculum Section first
section = soup.find('ul', id='digb_course_list')
courses=section.find_all('li', 'digb')
for course in courses:
    print(course.get_text().split('[')[0].split(' - ')[1].strip())
```

## Other option.enumerate

```python
res = requests.get('https://sy-netsci.github.io/DIGB241')
soup=bs(res.content, 'html.parser')

#Scraping Academic Curriculum Section first
section = soup.find('ul', id='digb_course_list')
courses=section.find_all('li', 'digb')
for index, course in enumerate(courses):
    print(str(index+1)+'.', course.get_text().split('[')[0].split(' - ')[1].strip())
```
