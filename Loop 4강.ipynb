## Loops in Python

## 1. For loops with Range

```python
for index in range(5):
    print(index)
```

## 2. For loops with items in List

```python
location = ['Seoul', 'Sejong', 'Daejeon', 'Pusan']
```

```python
for city in location:
    print(city)
```

## 3. For loops with items in Dict

```python
Dict = {'Name': 'Ben', 'Dept': 'DIGB', 'GPA': '4.0'}
```

```python
Dict.keys()
```

```python
for key in Dict.keys():
    print(key)
```

```python
Dict.values()
```

```python
for value in Dict.values():
    print(value)
```
