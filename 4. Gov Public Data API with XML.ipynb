- 전체 item중에서 원하는 정보만 뽑아봄
    - https://www.data.go.kr/tcs/dss/selectApiDataDetailView.do?publicDataPk=15043376
    - 출력결과(Response Element)에서 원하는 변수 살펴보기
- Data Manipulation
    - 'date:'
    - date.get_text()

````python
import requests
from bs4 import BeautifulSoup

key = '%2BpK691nADIPUQwJULL66bhGIJ0hxoNYpqGBhCKPD3UuOWt8yI9PXrOvDKbY2c1ERMh68TXVSq3yb%2BCaElZke0g%3D%3D'
params = '&pageNo=1&numOfRows=10&startCreateDt=20220310&endCreateDt=20220315'
gov_api = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?ServiceKey='+key+params

res = requests.get(gov_api)
soup = BeautifulSoup(res.content, 'html.parser')
mydata = soup.find_all('item')
for item in mydata:
    date = item.find('updatedt')
    death_toll = item.find('deathcnt')
    confirmed_cases = item.find('decidecnt')
    print('date:', date.get_text(), 'death_toll:', death_toll.get_text(), 'confirmed_cases:', confirmed_cases.get_text())
````

## 3. Save as an Excel file

````python
import requests
from bs4 import BeautifulSoup
import openpyxl

key = '%2BpK691nADIPUQwJULL66bhGIJ0hxoNYpqGBhCKPD3UuOWt8yI9PXrOvDKbY2c1ERMh68TXVSq3yb%2BCaElZke0g%3D%3D'
params = '&pageNo=1&numOfRows=10&startCreateDt=20220310&endCreateDt=20220315'
gov_api = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?ServiceKey='+key+params

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['A'].width = 30
excel_sheet.column_dimensions['B'].width = 30
excel_sheet.column_dimensions['C'].width = 30
excel_sheet.append(['Date', 'Death Toll', 'Confirmed Cases'])

res = requests.get(gov_api)
soup = BeautifulSoup(res.content, 'html.parser')
mydata = soup.find_all('item')
for item in mydata:
    date = item.find('updatedt')
    death_toll = item.find('deathcnt')
    confirmed_cases = item.find('decidecnt')
    excel_sheet.append([date.get_text(), death_toll.get_text(), confirmed_cases.get_text()])

excel_file.save('COVID19.xlsx')
excel_file.close()
````
