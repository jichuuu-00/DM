### 3-1. Inner Join
- <b>How to create a table that shows both info about students and their check-out history on April? (i.e., only for those who have check-out history)</b>

>- merge(df1, df2, how='inner')
    - the merge() method by default performs an inner join (how='inner')
>- Use intersection of keys from both frames
>- The output will contain all the records that have a mutual id in both df1 and df2
><img src="https://i.stack.imgur.com/YvuOa.png" width="350"/>

```python
pd.merge(student_df, checkout_april_df, on="Student_Id", how="inner")
#Sam and Will were deleted
```

>- How many times did Will borrow a book on April?

<b>NOTE.</b> Will borrowed a book and his info was deleted 

### 3-2. Left Outer Join
- <b>How to create a table that shows info about both those who have and do not have check-out history on April?</b>

>- merge(df1, df2, on='column', how='left')
>- Use keys from the left frame only
><img src="https://i.stack.imgur.com/BECid.png" width="350"/>

```python
pd.merge(student_df, checkout_april_df, on="Student_Id", how="left")
#Still shows info about Sam
```

### 3-3. Right Outer Join
>- merge(df1, df2, on='column', how='right')
>- Use keys from the right frame only
><img src="https://i.stack.imgur.com/8w1US.png" width="350"/>

```python
pd.merge(student_df, checkout_april_df, on="Student_Id", how="right")
#Book_Id=2 was borrowed by Will but the borrower info was deleted
```

### 3-4. Full Outer Join
>- merge(df1, df2, on='column', how='outer')
>- Use union of keys from both frames
><img src="https://i.stack.imgur.com/euLoe.png" width="350"/>

```python
pd.merge(student_df, checkout_april_df, on="Student_Id", how="outer")
```
