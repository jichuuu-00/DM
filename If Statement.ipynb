## Exercise 4.

* Print “You are eligible to graduate" if GPA is equal to or above 3.0 AND the TOEIC score is equal to or above 650

```python
GPA = input('What is your GPA?')
GPA_digit = float(GPA)

TOEIC = input('What is your TOEIC score?')
TOEIC_digit = int(TOEIC)

if GPA_digit > 3.0 and TOEIC_digit >= 650:
    print('You are eligible to graduate')
```

## If Else Statement

## Exercise 1.

* Print “Congratulations! You are eligible to graduate." if GPA is equal to or above 3.0 AND the TOEIC score is equal to or above 650
* if NOT satisfied, print "Keep up the good work!"

```python
GPA = input('What is your GPA?')
GPA_digit = float(GPA)

TOEIC = input('What is your TOEIC score?')
TOEIC_digit = int(TOEIC)

if GPA_digit > 3.0 and TOEIC_digit >= 650:
    print('You are eligible to graduate')
    
else:
    print('Keep up the good work!')
```

## Exercise 2.

* Print “Congratulations! You are eligible to graduate.” if the TOEIC score is equal to or above 650
* Print “Keep up the good work!" if the TOEIC score is below 650 and above 500
* Print “I’m sorry, you are NOT eligible to graduate” if the TOEIC score is equal to or below 500


```python
TOEIC = input('What is your TOEIC score?')
TOEIC_digit = int(TOEIC)

if TOEIC_digit >= 650:
    print('Congratulations! You are eligible to graduate')

elif 500 < TOEIC_digit < 650:
    print('Keep up the good work!')

else:
    print('I’m sorry, you are NOT eligible to graduate')
```

## Nested If Statement

## Exercise 1.

* if GPA >= 4.0 print "This applicant should be accpeted"

* Otherwise,
    - if GPA >= 3.5, print "This applicant should be given an in-person interview"
    - otherwise, print "This applicant should be rejected"


```python
GPA = input('What is your GPA?')
GPA_digit = float(GPA)

if GPA_digit >= 4.0:
    print("This applicant should be accpeted")

else:
    if GPA_digit >= 3.5:
        print("This applicant should be given an in-person interview")
    else:
        print("This applicant should be rejected")
```
