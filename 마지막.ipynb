```python
import openpyxl

def write_excel_template(filename, sheetname, item_lists):
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions['A'].width = 30
    
    if sheetname != '': #if it is blank, okay, if it has a certain name, i'll name it 'sheetname'
        excel_sheet.title = sheetname
    
    for item in item_lists:
        excel_sheet.append(item)
    
    excel_file.save(filename)
    excel_file.close()
```

```python
import requests
from bs4 import BeautifulSoup as bs

item_lists = list()

res = requests.get('https://finance.naver.com/sise/')
soup = bs(res.content, 'html.parser')
mydata = soup.select('ul.lst_major > li')
for item in mydata:
    stock_name = item.select_one('a')
    stock_price = item.select_one('span')
    stock_change = item.select_one('em > span')
    stock_info = [stock_name.get_text(), stock_price.get_text(), stock_change.get_text()]
    item_lists.append(stock_info)
```

```python
print(item_lists)
```

```python
write_excel_template('Naver_Stock.xlsx', 'Stock', item_lists)
```
